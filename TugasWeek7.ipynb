```python
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification

# Using pipeline
pipe = pipeline("text-classification", model="mrm8488/distilroberta-finetuned-financial-news-sentiment-analysis")

# Load model and tokenizer directly
tokenizer = AutoTokenizer.from_pretrained("mrm8488/distilroberta-finetuned-financial-news-sentiment-analysis")
model = AutoModelForSequenceClassification.from_pretrained("mrm8488/distilroberta-finetuned-financial-news-sentiment-analysis")
```

```python
# Sample sentence
sentence = "The company announced record-breaking profits for this quarter."

# Using pipeline for prediction
result = pipe(sentence)
print(result)
```

```python
# Sample synthetic dataset (replace with actual data for better accuracy)
synthetic_data = [
    {"text": "The company's stock soared after the earnings report.", "label": "positive"},
    {"text": "There were no significant changes in the market today.", "label": "neutral"},
    {"text": "The company faced major losses in the last quarter.", "label": "negative"},
]

# Evaluate model on the synthetic dataset
correct = 0

for item in synthetic_data:
    result = pipe(item['text'])[0]['label']
    if result.lower() == item['label']:
        correct += 1

accuracy = correct / len(synthetic_data)
print(f"Accuracy on synthetic dataset: {accuracy * 100:.2f}%")
```

```python
from transformers import BertModel

# Load a BERT-like model for attention visualization
model_with_attention = AutoModelForSequenceClassification.from_pretrained("mrm8488/distilroberta-finetuned-financial-news-sentiment-analysis", output_attentions=True)

# Encode the sentence
inputs = tokenizer(sentence, return_tensors="pt")
outputs = model_with_attention(**inputs)

# Get attentions
attentions = outputs.attentions
print("Attention layers shape:", len(attentions), attentions[0].shape)
```
